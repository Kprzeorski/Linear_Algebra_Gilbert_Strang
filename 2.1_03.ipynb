```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np
%matplotlib
```

```python
# 1x+1y+0z=5 & 0x+0y+1z=4 overlayed on the planes in the previous problems

fig = plt.figure(figsize=(12,12))
ax = fig.gca(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.set_xlim3d(-5,16)
ax.set_ylim3d(-5,16)
ax.set_zlim3d(-5,16)
mpl.rcParams['legend.fontsize'] = 20

# DRAW IN 'ROW SPACE'
# Equation 1
Z = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
Y, Z = np.meshgrid(Y, Z)
X = 2 - (0*-Y + 0*-Z)
surf = ax.plot_surface(X, Y, Z, linewidth=0,  alpha=.1)
# Equation 2
X2 = np.arange(-5, 5, 0.25)
Z2 = np.arange(-5, 5, 0.25)
X2, Z2 = np.meshgrid(X2, Z2)
Y2 = 3-(0*-X2 + 0*-Z2)
surf2 = ax.plot_surface(X2, Y2, Z2, linewidth=0, alpha=.1)
# Equation 3
X3 = np.arange(-5, 5, 0.25)
Y3 = np.arange(-5, 5, 0.25)
X3, Y3 = np.meshgrid(X3, Y3)
Z3 = 4-(0*-X3 + 0*-Y3)
surf3 = ax.plot_surface(X3, Y3, Z3, linewidth=0, alpha=.1)

Z4 = np.arange(-2.5, 5, 0.25)
Y4 = np.arange(-2.5, 5, 0.25)
Y4, Z4 = np.meshgrid(Y4, Z4)
X4 = (5 - 1*Y4 - 0*Z4) #1x+1y+0z=5
surf4 = ax.plot_surface(X4, Y4, Z4, linewidth=0,  alpha=.9)
ax.text3D(10,0,0,"1x+1y+0z=5 - dark plane")
ax.quiver3D(10,0,0,-5,0,0,arrow_length_ratio=0,color='k')

# Equation 5 = Equation 2 * 3
X5 = np.arange(-2.5, 5, 0.25)
Y5 = np.arange(-2.5, 5, 0.25)
X5, Y5 = np.meshgrid(X5, Y5)
Z5 = (4 - 0*X5 - 0*Y5) #0x+0y+1z=4
surf5 = ax.plot_surface(X5, Y5, Z5, linewidth=0, alpha=.9)
ax.text3D(0,0,10,"0x+0y+1z=4 - dark plane")
ax.quiver3D(0,0,10,0,0,-6,arrow_length_ratio=0,color='k')

# DRAW IN 'COLUMN SPACE'
ax.quiver3D(0,0,0,2,0,0,color='g')
ax.quiver3D(2,0,0,0,3,0,color='g')
ax.quiver3D(2,3,0,0,0,4,color='g')
ax.quiver3D(0,0,0,2,3,4,arrow_length_ratio=.1)
ax.scatter3D(2,3,4,s=100)
labels = ['column1', 'columns1+2', 'columns1+2+3/ vector x/X']
for x_label, y_label, z_label, label in zip([2,2,2], [0,3,3], [0,0,4], labels):
    ax.text(x_label, y_label, z_label, label)
```
