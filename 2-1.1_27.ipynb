## Row Picture

```python

fig = plt.figure(figsize=(12,12))
ax = fig.gca(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim3d(-15,15)
ax.set_ylim3d(-15,15)
ax.set_zlim3d(-15,15)
mpl.rcParams['legend.fontsize'] = 20

# to visualize solution on a line, let us arbitrarily choose 2 equations:
# 2 equations, 3 unknowns
# x + 2y +z = 0
# x + y -5z = 6
# express equations in terms of z:
# x = 12+11z
# y = -6- 6z
# z = z
# E.G. z = -1 then x = 1, y = 0 :
ax.text3D(1,0,-1,"x1")
point=ax.scatter3D(1,0,-1,s=150, marker='o',c='r')
# another test point... run program and see dots producing an imaginary line?
# solutions reside on this line (where the planes intersect)
ax.text3D(-10,6,-2,"x2")
point=ax.scatter3D(-10,6,-2,s=150, marker='o',c='r')
ax.text3D(-4.5,3,-1.5,"x2")
point=ax.scatter3D(-4.5,3,-1.5,s=150, marker='o',c='r')

# 2 equations, 3 unknowns
X = np.arange(-15, 15, 0.25)
Y = np.arange(-15, 15, 0.25)
X, Y = np.meshgrid(X, Y)
Z = ( -X - 2*Y) #x+2y+z=0
surf = ax.plot_surface(X, Y, Z, linewidth=0, alpha=.2)

X2 = np.arange(-15, 15, 0.25)
Y2 = np.arange(-15, 15, 0.25)
X2, Y2 = np.meshgrid(X2, Y2)
Z2 = (6 -X2 - Y2)/-5 #x+y-5z=6
surf = ax.plot_surface(X2, Y2, Z2, linewidth=0, alpha=.2)
```
